# merge()

```python
# sử dụng merge() để join 2 hoặc nhiều bảng dữ liệu
left = pd.DataFrame({"key": ["A", "B", "C"], "lval": [1, 2, 99]})
display(left)

right = pd.DataFrame({"key": ["A", "B", "F"], "rval": [4, 5, 6]})
display(right)
```

```python
pd.merge(left, right, on="key", how = "inner")
```

```python
pd.merge(left, right, on="key", how = "left")
```

```python
pd.merge(left, right, on="key", how = "right")
```

```python
pd.merge(left, right, on = "key", how = "outer")
```

# Pivot table

```python
import numpy as np

df = pd.DataFrame(
    {
        "A": ["A", "B", "C"] * 4,
        "B": ["Tokyo", "Chiaba"] * 6,
        "C": np.random.randn(12),
        "D": np.arange(12),
    }
)

display(df)

pd.pivot_table(df, values=["C", "D"], index=["A"], columns=["B"])
```

# Chuỗi thời gian

```python
# Tạo một chuỗi thời gian
date_range = [f"2024080{str(i)}" for i in np.arange(1,7)]

# Tạo DataFrame với chuỗi thời gian
df = pd.DataFrame({'date': date_range, 'value': range(6)})
display(df)
```

```python
# # Chuyển đổi cột 'date' thành định dạng datetime
df["date"] = pd.to_datetime(df["date"], format = "%Y%m%d")
df
```

```python
df.dtypes
```

```python
# Trích xuất năm, tháng, ngày từ cột 'date'
df["year"] = df["date"].dt.year
df["month"] = df["date"].dt.month
df["day"] = df["date"].dt.day
display(df)
```

```python
# strftime
df["formated_date"] = df["date"].dt.strftime("%d/%m/%y (%A)")
```

```python
display(df)
```
