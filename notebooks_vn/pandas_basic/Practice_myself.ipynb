# Dữ liệu bị thiếu

```python
# Đọc dữ liệu tử bảng product
product_path = "../../data/preprocessed/product.csv"
dtype = {
    'product_cd': str,
    'category_major_cd': int,
    'category_medium_cd': int,
    'category_small_cd': int,
    'unit_price': float,
    'unit_cost': float
}
product_df = pd.read_csv(product_path, dtype = dtype)
product_df.head()
```

```python
product_df.info()
```

```python
product_df.shape
```

```python
product_df.isna().sum().to_frame()
```

## Lọc ra những hàng bị missing

```python
product_df[product_df.isna().any(axis = 1)]
```

## Sử dụng hàm fill() để điền các giá trị missing

```python
mean_unit_price = product_df["unit_price"].mean()
print(f"Giá trị trung bình của cột unit price là: {mean_unit_price}")
```

```python
product_df.loc[:, "unit_price"] = product_df["unit_price"].fillna(mean_unit_price)
product_df.isna().sum().to_frame()
```

```python
product_df[product_df.isna().any(axis = 1)]
```

```python
mean_unit_cost = product_df["unit_cost"].mean()
print(mean_unit_cost)
```

```python
# fill missing value in unit_cost columns
product_df.loc[:, "unit_cost"] = product_df["unit_cost"].fillna(mean_unit_cost)
product_df.isna().sum().to_frame()
```

# Nối theo chiều ngang - Merge() function

```python
# sử dụng merge() để join 2 hoặc nhiều bảng dữ liệu
left = pd.DataFrame({"key":["A", "B", "C"], "lval": [1,2,99]})
right = pd.DataFrame({"key": ["A", "B", "F"], "rval": [4,5,6]})

display(left)
display(right)
```

## Inner join

```python
pd.merge(left, right, on = "key", how = "inner")
```

## Left join

```python
pd.merge(left, right, on = "key", how = "left")
```

## Right join

```python
pd.merge(left, right, on = "key", how = "right")
```

## Outer join

```python
pd.merge(left, right, on = "key", how = "outer")
```

# Chuỗi thời gian

```python
# Tạo một chuỗi thời gian
date_range = [f"2024080{str(i)}" for i in np.arange(1,7)]

# Tạo DataFrame với chuỗi thời gian
ts_df = pd.DataFrame({'date': date_range, 'value': range(6)})
display(ts_df)
```

```python
ts_df.info()
```

## Chuyển đổi cột 'date' thành định dạng datetime

```python
ts_df["date"] = pd.to_datetime(ts_df["date"], format = "%Y%m%d")

ts_df.dtypes
```

## Trích xuất ngày tháng năm từ cột date

```python
ts_df["year"] = ts_df["date"].dt.year
ts_df["month"] = ts_df["date"].dt.month
ts_df["day"] = ts_df["date"].dt.day

ts_df
```

## Định dạng kiểu chuỗi thời gian

```python
# Định dạng chuỗi thời gian thành chuỗi ký tự theo định dạng 'YYYY-MM-DD'
ts_df['formatted_date'] = ts_df['date'].dt.strftime('%Y/%m/%d')
display(ts_df)
```

```python
# Định dạng chuỗi thời gian thành chuỗi ký tự với các thành phần khác nhau
ts_df['formatted_date_full'] = ts_df['date'].dt.strftime('%A, %d %B %Y')
display(ts_df)
```

```python
# Định dạng chuỗi thời gian thành chuỗi ký tự theo định dạng tùy chỉnh 'DD/MM/YYYY'
ts_df['custom_formatted_date'] = ts_df['date'].dt.strftime('%d/%m/%Y')
display(ts_df)
```

Các định dạng thông dụng sử dụng trong strftime:

- %Y: Năm đầy đủ (ví dụ: 2024)
- %m: Tháng (01 đến 12)
- %d: Ngày (01 đến 31)
- %H: Giờ (00 đến 23)
- %M: Phút (00 đến 59)
- %S: Giây (00 đến 59)
- %A: Tên đầy đủ của ngày trong tuần (ví dụ: Monday)
- %B: Tên đầy đủ của tháng (ví dụ: January)
